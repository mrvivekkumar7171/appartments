- `ydata_profiling` : it is an open-source Python library used to automate and streamline the Exploratory Data Analysis (EDA) process. It generates comprehensive, interactive HTML or JSON reports for pandas or Spark DataFrames with a single function call, providing detailed statistics, visualizations, and data quality alerts.
- `Sweetviz`: Focuses on high-density visualizations and is widely considered the best tool for comparing two datasets (e.g., training vs. testing data) side-by-side. It generates a self-contained, interactive HTML report with just two lines of code.
- `D-Tale`: Provides an interactive web-based GUI that acts like a powerful spreadsheet within your browser or notebook. Unlike static reports, D-Tale allows you to filter, sort, and even perform basic data cleaning and feature engineering directly through its interface without writing additional code.

```python
from ydata_profiling import ProfileReport
import pandas as pd

# Load your dataset
df = pd.read_csv('A:/CODES/PROJECTS/appartments/Data/gurgaon/properties_cleaned_v2.csv').drop_duplicates()

# Create the ProfileReport object
profile = ProfileReport(df, title='Ydata Profiling Report', explorative=True)

# Generate the report
profile.to_file("A:/CODES/PROJECTS/appartments/EDA/output_report.html")
```
